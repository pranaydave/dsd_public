<h2>Data preperation</h2>
<p>In this jupyternotebook, we prepare data to predict youtube video clickthrough rate using thumbnail
<p>You should download YouTube analytics data in CSV
<p>The code here will read the video id (content) from youtube analytics data, get thumbnail image and save it locally

```python
import requests
from PIL import Image
from io import BytesIO
import pandas as pd
```

<b>Directory for storing thumbnail and to read youtube channel data</b>

```python
TN_DATA_DIR = 'yt_thumbnail' #We will save thumbnail in this directory
YT_STAT_DIR = 'yt_youtube' #This directory has youtube statistics data
```


<b>Here we read youtube analytics data in dataframe</b>
<b>The content field is the video-id. It is used to make URL of the thumbnail, which we can fetch from Youtube</b>


```python
file_in = YT_STAT_DIR+'/yt_channel_data.csv'
```

```python
df = pd.read_csv(file_in)
```

```python
df.head(5)
```

```python
# Remove rows where 'ctr' is 0.0 or NaN
df1 = df[(df['Impressions click-through rate (%)'] != 0.0) & (df['Impressions click-through rate (%)'].notna())]
```

<b>Gettng thumnbail image and storing it locally</b>
<p> Not all images wil be successfully retreived. So we create a list which has only images which are successfully retreived

```python
def get_image(content,url):
    
    # Fetch the image
    response = requests.get(url)
    if response.status_code == 200:
        image = Image.open(BytesIO(response.content))

        # Save the image
        file_out = TN_DATA_DIR+'/'+content+'.jpg'
        image.save(file_out)
        print(content, " Image saved successfully!")
        return True
    else:
        print(content," Failed to download the image")
        return False
```

```python
lst_items = []
lst_header = [['content','image_name','url','video_title','ctr']]
for index, row in df1.iterrows():
    content = row['Content']
    url = 'https://img.youtube.com/vi/'+content+'/hqdefault.jpg'
    ctr = row['Impressions click-through rate (%)']
    video_title = row['Video title']
    image_name = content+'.jpg'
    saved = get_image(content,url)
    print (saved)
    if saved == True:
        lst_items.append([content,image_name,url,video_title,ctr])
    #print (content, url,ctr)
```
